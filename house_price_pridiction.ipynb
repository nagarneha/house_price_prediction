# Target Variable analysis (SalePrice)

```python

plt.figure(figsize=(8,5))
sns.boxplot(x=train['SalePrice'])
plt.title("Boxplot of SalePrice")
plt.show()
```

# correlation with saleprice (numerical features)

```python
plt.figure(figsize=(10,8))
corr = train.select_dtypes(include=['int64', 'float64']).corr()   # only numeric
sns.heatmap(corr[['SalePrice']].sort_values(by='SalePrice', ascending=False), 
            annot=True, cmap="coolwarm")
plt.title("Correlation of Numerical Features with SalePrice")
plt.show()
```

# Importent Numnerical Features VS Saleprice

```python
features = ['GrLivArea', 'TotalBsmtSF', 'GarageArea', 'YearBuilt']
for col in features:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=train[col], y=train['SalePrice'])
    plt.title(f"{col} vs SalePrice")
    plt.show()
```

#Categorical Features VS SalePrice

```python
categorical = ['Neighborhood', 'OverallQual', 'HouseStyle', 'MSZoning']

for col in categorical:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=col, y='SalePrice', data=train)
    plt.xticks(rotation=45)
    plt.title(f"{col} vs SalePrice")
    plt.show()
```

# Time-based Features

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='YrSold', y='SalePrice', data=train)
plt.title("SalePrice Trend over Years")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='YearBuilt', y='SalePrice', data=train)
plt.title("YearBuilt vs SalePrice")
plt.show()
```

```python
# Remodeling age (kab last time modify hua)
train['RemodAge'] = train['YrSold'] - train['YearRemodAdd']

# Total bathrooms
train['TotalBath'] = (train['FullBath'] + 0.5*train['HalfBath'] + 
                      train['BsmtFullBath'] + 0.5*train['BsmtHalfBath'])

# Total square footage
train['TotalSF'] = train['GrLivArea'] + train['TotalBsmtSF']
```
